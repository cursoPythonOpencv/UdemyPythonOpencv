# Segmentao de um objeto por espao de cor HSV

### Objetivos:

* Como converter imagens para um espao de cores HSV (BGR <-> HSV)
* Segmentar um objeto de um video

### Espao de cor HSV

* Um dos modelos principais e mais utilizados para representao de cor. 
* Permite um grande grau de separao entre a cor e a iluminao. 
* So os canais para representar a matiz (hue), a saturao (saturation) e o valor (value).



* O componente Value  desagregado da informao de cor, enquanto o Hue e o Saturation esto intimamente relacionados  percepo humana de cores. 

### Como encontrar os valores HSV de uma imagem:

 muito simples e voc pode usar a mesma funo, **cv2.cvtColor()**


```python
import numpy as np
import cv2

#Passar padro BGR
azul = np.uint8([[[255, 0, 0]]])
hsv_azul = cv2.cvtColor(azul, cv2.COLOR_BGR2HSV)

verde = np.uint8([[[0, 255, 0]]])
hsv_verde = cv2.cvtColor(verde, cv2.COLOR_BGR2HSV)

vermelho = np.uint8([[[0, 0, 255]]])
hsv_vermelho = cv2.cvtColor(vermelho, cv2.COLOR_BGR2HSV)

print(f"A cor azul no espao HSV possui o valor {hsv_azul}" )
print(f"A cor verde no espao HSV possui o valor {hsv_verde}" )
print(f"A cor vermelho no espao HSV possui o valor {hsv_vermelho}" )
```

### Limites superiores e inferiores

* limite inferior: [H-20, 100,100] 
* limite superior: [H+20, 255, 255]  

* ferramenta de edio de imagem como o **GIMP** ou qualquer **conversor online** 

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()
    hsv =  cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lim_inf = np.array([100, 100, 100])
    lim_sup = np.array([140, 255, 255])
    
    color_mask = cv2.inRange(hsv, lim_inf, lim_sup)

    (couts,hir) = cv2.findContours(color_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    for cout in couts:
        area = cv2.contourArea(cout)

        if (area > 800):
            x,y,w,h = cv2.boundingRect(cout)
            frame = cv2.rectangle(frame, (x,y),(x+w,y+h),(0,0,255),2)
            
    
            
    cv2.imshow('Mascara', color_mask)
    cv2.imshow('ObjectDetctionTrack', frame)

    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
cap.release()    
```
